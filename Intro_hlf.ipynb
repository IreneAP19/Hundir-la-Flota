```python


#Creo el tablero de juego 10x10

#return np.full((filsd,columnas), "_") = que hacer la de zeros
def crear_tablero(filas=10,columnas=10):
    tablero= np.zeros((filas,columnas), dtype = str)#hago una tupla para que sea bidimensional
    #hago un array de zeros e indico que sea de tipo string para que al sustituir un 0=int por un "_"=str no me de problemas
    tablero[:]="_"#sustituyo todos los elementos del tablero(0) por "_"

    return tablero
        
```

Colocamos los barcos:
1- necesito las variables de barco(cada uno con su respectiva medida) y el tablero donde voy a colocarlo

3- en el tablero que este usando sustitullo en la coordenada del barco el _ por un barquito

```python
#FUNCION PARA COLOCAR LOS BARCOS SEGUN SU POSICION
'''vamos a llamar a barco junto a la funcion que nos genera las posiciones'''
import emoji
def colocar_barcos(barco,tablero):
    '''cada barco va atener uno/varios valor/es en columna y en fila
       lo que queremos es sustituir dichos valores en el tablero de juego
       Para ello vamos a crear un bucle for(hay una medida determinada en cada uno pero varia segun el barco)
        
            para cada valor de columna y de fila que encuentres en la variable barco:
                coge el tablero y en la posicion (fila,columna) cambia"_" por un barquito
                devuelveme el nuevo tablero'''
    for fila,columna in barco:
        tablero[fila,columna] = emoji.emojize(':ship:') #no me va a dar fallo pq estamos trabajando con str
    return tablero #devuelveme el tablero con los cambios
    
#barco sera una lista de posiciones y el tablero va a ir variando entre el mio y el del pc
```

Para crear el barco de forma aleatoria, que sera el que usemos cuando ponga el pc sus barcos:<br>

    Necesito una lista vacia donde voy a poner todas las posiciones de mis barcos.<br>

    Como no quiero que se me repitan hago un conjunto donde voy a guardar las posiciones que vaya cogiendo en cada iteracion<br>
        -> para cada medida que tengo en la eslora:<br>
            ponemos WHILE TRUE ya que con este tipo de iteraciones podemos encontrarnos con que no obtengamos un barco valido(que se salga del tablero, que repita posicion..) por ello este comando nos permite repetir el bucle hasta que de con un barcon con posiciones validas(buscado)<br>

                - necesitamos ir almacenando las posiciones de cada barco por individual, para ello creamos una lista dentro del 
                     bucle ya que queremos cogerlas de una en una 
                - inicializamos dos valores random (i,j) que se actualizaran en cada repeticion del bucle (para cada barco) 
                - del mismo modo tenemos que considerar que la orientacion tambien puede variar de forma aleatoria

                - Con estas condiciones aleatorias hacemos un estudio de casos considerando las dimensiones del tablero
                        en la lista de barco_actual vamos a침adiendo tantas posiciones como larga sea la eslora en cada caso(medida)

                - La unica condicion que queda por revisar es que los valores tomados no esten ya en otro barco. 
                    Para ello hacemos un :
                        si todas las posiciones  del barco actual no estan en el conjunto de posiciones ocupadas a침adelas 
                        (barco_actual) a la lsita de barcos y metelas en el set de ocupadas para que se tengan en cuenta para la 
                        siguiente iteracion

                    rompe el bucle y vuelve a hacer lo mismo hasta que no tengamos mas medidas en esloras



```python
import random
def crear_barco(eslora): #eslora=len(barco)
    #esta lista de barcos es la que voy a llamar en colocar barcos
    barcos = []#inicializo una lista para guardar todas las posiciones de los barcos
        
    ocupado = set() #conjunto de casillas ya ocupadas, las almacenamos para evitar que una posicion pise a otra
    for medida in eslora:
        while True:#mientras que exista la posicion   
            barco_actual =[]#aqui almacenamos las posiciones de cada barco de forma individual
               #Le pido que me de dos valores iniciales para cada eslora
            i= np.random.randint(0,9)
            j = np.random.randint(0,9)
            #tambien tengo que considerar que la direccion puede ser aleatoria
            orientacion = random.choice(["vertical","horizontal"])  #me da de forma aleatoria la orientacion
            
            

        #estudio de casos segun la orinetacion y contemplando la dimension del tablero y longitud de cada eslora
            if orientacion == "vertical" and (i+ medida <=9):
                barco_actual= [(i+n, j) for n in range(medida)]#n=0,...,num eslora. Por eso no ponemos el primer valor en barco antes del if
                
            elif orientacion == "vertical" and  (i- medida>=0):
                barco_actual = [(i-n, j) for n in range(medida)]
                
        
        
            elif orientacion == "horizontal" and (j + medida <=9):
                barco_actual= [(i, j+ n) for n in range(medida)]
                
            elif orientacion == "horizontal" and  (j- medida >=0):
                barco_actual = [(i, j-n) for n in range(medida)]
            
            #comprobamos que la posicion del barco no esta ocupada, si no hacemos un break y volvemos al bucle
            if all(pos not in ocupado for pos in barco_actual):
                barcos.append(barco_actual)
                ocupado.update(barco_actual)#pon esa posicion como ocupada para que no se pisen
                break# haz el bucle con una nueva medida de eslora
            
            
    return barcos#mi lista de posiciones de cada barco
            
```

Para disparar:<br>
1- necesitamos una casilla(aleatoria si es de pc o introducida por el jugador) y un tablro donde se dipare<br>
2- la casilla cuenta con un num de fila y otro de columna<br>
3- si la casilla es un _ cambiala por una gota (AGUA). En otro caso le habremos dado al barco y la cambiamos por una bomba (TOCADO)<br>
4- devuelveme el tablero actualizado

```python
import emoji
def disparar(casilla,tablero):
    for fila,columna in casilla:
        if tablero[fila,columna] == "_":
            
            tablero[fila,columna] = emoji.emojize(':droplet:')
            return "Agua"
        else:
            tablero[fila,columna] = emoji.emojize(':bomb:')
            return "Tocado"
        
    return tablero
```

    3- una vez creadas las posiciones (con las medidas correspondientes de cada barco) hay que colocarlos en el tablero. Para ello usamos la funcion colocar_barco. Para esta funcion necesitamos una posicion y un tablero(en cada posicion pone un barquito)
